# Other Models

**Mean Reversion model** - This trading model is based on the idea that the exchange rate will eventually return to its mean value. You can use this model to buy USD when the KES is overvalued and sell USD when the KES is undervalued.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from sklearn.metrics import mean_squared_error
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'Dataset/Key CBK Indicative Exchange Rates.csv'
exchange_rates_data = pd.read_csv(file_path)

# Assuming 'Date' and 'Mean' columns are relevant
time_series_data = exchange_rates_data[['Date', 'Mean']]

# Convert 'Date' to datetime
time_series_data['Date'] = pd.to_datetime(time_series_data['Date'], format='%d/%m/%Y')

# Set 'Date' as the index
time_series_data.set_index('Date', inplace=True)

# Define a rolling window size (adjust as needed)
window_size = 30

# Calculate the rolling mean
time_series_data['RollingMean'] = time_series_data['Mean'].rolling(window=window_size).mean()

# Plot the original data and the rolling mean
plt.figure(figsize=(12, 6))
plt.plot(time_series_data['Mean'], label='Exchange Rate')
plt.plot(time_series_data['RollingMean'], label=f'Rolling Mean ({window_size} days)', color='red')
plt.xlabel('Date')
plt.ylabel('Exchange Rate')
plt.title('Mean Reversion Strategy')
plt.legend()
plt.show()
```

The mean reversion model graph shows the exchange rate between the Kenyan Shilling (KES) and the US Dollar (USD) over time. The blue line is the actual exchange rate, and the red line is the predicted exchange rate using the mean reversion model.

The mean reversion model is a statistical model that predicts that the exchange rate will eventually return to its mean value. The model is based on the idea that the exchange rate cannot stay overvalued or undervalued forever.

The graph shows that the mean reversion model is generally able to track the overall trend of the exchange rate reasonably well. However, there are some periods where the predicted exchange rate deviates significantly from the actual exchange rate. This is likely due to factors that are not captured by the mean reversion model, such as news events and economic shocks.

Overall, the graph suggests that the mean reversion model can be used to predict the exchange rate with a reasonable degree of accuracy. However, it is important to note that the mean reversion model is only a statistical model, and it cannot perfectly predict the future.

Here are some specific observations about the graph:

* The predicted exchange rate is generally higher than the actual exchange rate in the first half of the graph. This suggests that the mean reversion model is overestimating the exchange rate during this period.
* The predicted exchange rate is generally lower than the actual exchange rate in the second half of the graph. This suggests that the mean reversion model is underestimating the exchange rate during this period.
* The predicted exchange rate is closest to the actual exchange rate around the middle of the graph. This suggests that the mean reversion model is performing best during this period.

It is important to note that the mean reversion model is a forecasting model, not a trading model. This means that the mean reversion model cannot be used to guarantee profits in the foreign exchange market.

**Trading strategy**

One way to use the mean reversion model to trade the exchange rate is to buy USD when the KES is overvalued and sell USD when the KES is undervalued.

To determine whether the KES is overvalued or undervalued, you can look at the difference between the actual exchange rate and the predicted exchange rate. If the actual exchange rate is significantly higher than the predicted exchange rate, then the KES is overvalued. Conversely, if the actual exchange rate is significantly lower than the predicted exchange rate, then the KES is undervalued.

Once you have determined whether the KES is overvalued or undervalued, you can place a trade accordingly. For example, if you believe that the KES is overvalued, you can buy USD. When the KES returns to its mean value, you can then sell the USD for a profit.

It is important to note that the mean reversion model is not a perfect trading strategy. There will be times when the KES does not revert to its mean value, and you may lose money on your trade. Therefore, it is important to use risk management techniques, such as stop-loss orders, to limit your losses.
